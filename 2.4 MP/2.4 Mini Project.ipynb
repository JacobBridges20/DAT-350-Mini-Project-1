#1 Percentage of the ages of the repsondents to the survey

```python
# function that gives me the percentage of each age group in the dataset

def age_group_percentage():
    import csv

    with open('DAT_350_MP_1.csv') as csvfile:
        openCSV = csv.reader(csvfile)
        next(openCSV)  # skip header

        age_groups = {
            "10-20": 0,
            "21-30": 0,
            "31-40": 0,
            "41-50": 0,
            "51-60": 0,
            "61-70": 0,
            "71-80": 0,
            "81-90": 0,
            "91-100": 0
        }

        total_count = 0

        for row in openCSV:
            age_group = row[1].strip()

            if age_group in age_groups:
                age_groups[age_group] += 1
                total_count += 1

        percentages = {}
        for group, count in age_groups.items():
            percentages[group] = (count / total_count * 100) if total_count > 0 else 0

        return percentages

print(age_group_percentage())
```
